```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
import matplotlib.pyplot as plt
fl = pd.read_csv('../csv/homeprice.csv')
print(fl)
fl.describe()
```

```python
# %matplotlib inline
plt.xlabel("area")
plt.ylabel("price")
plt.scatter(fl.area,fl.price,color="red", marker='x')
plt.plot(fl.area,fl.price)
plt.grid()
```

```python
new_df=fl.drop('price',axis="columns")
new_df
```

```python
model = LinearRegression()
```

```python
model.fit(new_df.values,fl.price)
```

```python
model.score(new_df.values,fl.price)
```
