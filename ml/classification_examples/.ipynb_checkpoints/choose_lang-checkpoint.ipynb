```python
import pandas as pd
```

```python
df = pd.read_csv('../csv/choose_lang.csv')
df.head()
```

```python
df.groupby('lang').describe()
```

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(df.quesetion,df.lang, random_state=43)
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
v = CountVectorizer()
```

```python
x_train_count = v.fit_transform(x_train)
x_train_count.toarray()
```

```python
from sklearn.naive_bayes import MultinomialNB
```

```python
model = MultinomialNB()
```

```python
model.fit(x_train_count, y_train.values)
```

```python
model.score(v.transform(x_test),y_test.values)
```

```python
model.predict(v.transform(['comment in css']))
```

```python
def createLabel(lang):
    if lang==1:
        return 'Html'
    elif lang==2:
        return 'Bootstrap'
    elif lang==3:
        return 'Js'
    else:
        return 'Css'
```

```python
df['label'] = df.lang.apply(createLabel)
df.head()
```

```python
df.drop(['Unnamed: 2'],axis='columns', inplace=True)
```
