```python
import pandas as pd
area = pd.read_csv('area.csv')
```

```python
area.describe()
```

```python
dummy = pd.get_dummies(area.town)
dummy.astype(int)
```

```python
df_columns = pd.concat([area,dummy.astype(int)],axis="columns")
df_columns
```

```python
df_columns.drop(['town','c'],axis="columns",inplace=True)
```

```python
df_columns
```

```python
x = df_columns.drop(['price'],axis="columns")
```

```python
x
```

```python
y = df_columns.price
```

```python
y
```

```python
from sklearn.linear_model import LinearRegression
```

```python
model = LinearRegression()
model.fit(x,y)
```

```python
model.predict([[3200,0,1]])
```
