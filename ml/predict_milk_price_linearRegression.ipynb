```python
import pandas as pd
from sklearn.linear_model import LinearRegression
```

```python
df = pd.read_csv('milk_price.csv')
df
```

```python
df.describe()
```

```python
dummy_data = pd.get_dummies(df.MilkGrade)
dummy_data.astype(int)
```

```python
df_new_datas = pd.concat([df,dummy_data.astype(int)],axis="columns")
df_new_datas
```

```python
df_new_datas.drop(['MilkGrade','C'],axis="columns",inplace=True)
```

```python
df_new_datas
```

```python
x = df_new_datas.drop(['price'],axis="columns")
x
```

```python
y=df_new_datas.price
```

```python
y
```

```python
model = LinearRegression()
model.fit(x.values,y.values)
```

```python
model.predict([[100,0,1]])
```

```python
model.score(x.values,y)
```
