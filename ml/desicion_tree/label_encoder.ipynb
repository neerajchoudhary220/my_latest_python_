```python
import pandas as pd
df = pd.read_csv('../csv/laptop.csv')
df
```

```python
from sklearn.preprocessing import LabelEncoder
```

```python
df['Laptop']=LabelEncoder().fit_transform(df.laptop)
df['Ram']=LabelEncoder().fit_transform(df.ram)
df['Speed']=LabelEncoder().fit_transform(df.speed)
df
```

```python
new_df = df.reindex(columns=['laptop','Laptop','ram','Ram','speed','Speed'])
```

```python
new_df
```
